```python
from google.colab import auth
auth.authenticate_user()

import gspread
from google.auth import default
creds, _ = default()

gc = gspread.authorize(creds)

worksheet = gc.open('[TDKR_ RIOT GAMES] Attendance Tracker').worksheet("Attendance")
# worksheet = gc.open('[TDKR_ RIOT GAMES] Attendance Tracker').worksheet("Billable")

# get_all_values gives a list of rows.
df = worksheet.get_all_values()
print(df)

# Convert to a DataFrame and render.
import pandas as pd


df = pd.DataFrame.from_records(df)
```

```python
new_header = df.iloc[1] #grab the first row for the header
df = df.iloc[2:] #take the data less the header row
df.columns = new_header #set the header row as the df header
```

```python
df.tail()
```

```python
df3 = df.iloc[:,7:].drop(columns=['Riot Email','Phone number','VPN ID','VPN PW','Team'])
```

```python
df4 = pd.DataFrame((df3.set_index(['TDCX email'])).stack()).reset_index()
```

```python
df5= df4.rename(columns={1:'Date',0:'Status' })
df5.head(2)
```

```python
df6= df[['Korean Name', 'Position', 'Batch', 'TDCX email']]
```

```python
df7 = pd.merge(df6,df5, how='right',on='TDCX email')
```

```python
df7['Date'] = pd.to_datetime(df7['Date'])
df8 = df7.dropna(subset=['Status'])
df8.columns.to_list()
```

```python
df8.tail(10)
```

```python
df9 = df8[(df8['TDCX email']!= "") & (df8['Status']!= "")]
```
